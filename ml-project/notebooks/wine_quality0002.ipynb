# Residual Sugar 

```python
wine['residual sugar'].describe()
```

```python
wine['residual sugar'].apply(np.log).plot.hist()
```

```python
duration = wine['residual sugar'].apply(np.log)

q3 = duration.quantile(0.75)
q1 = duration.quantile(0.25)

IQR = q3 - q1 

a = (duration >= q1 - 1.5 * IQR) & (duration <= q3 + 1.5 *IQR)
duration.loc[a].quantile(1)
```

```python
duration[duration < 1.410986973710262].plot.hist()
```
